## Dates
The following attempts to convert dates into a standard YYYY-MM-DD format

```python
netflix['computer_date'] = pd.to_datetime(netflix['date_added'])
netflix['computer_date']
```

### Handling NAs
We can simply drop them (note, 10 fewer observations)

```python
netflix['computer_date'].dropna()
```

```python
from datetime import datetime as dt
```

Assumes all NAs as 2020-01-01

```python
netflix['computer_data'].fillna(dt(2020, 1, 1), inplace=True)
```

### Splitting Data

```python
netflix[['Date Part 1', 'Date Part 2']] = netflix['date_added'].str.split(', ',expand=True)
netflix
```

### Renaming Columns

```python
# This utilizes a dictionary, where the key is the original name and the value is the new name
netflix.rename(columns={'Date Part 2': 'Year'}, inplace=True)
netflix
```

```python
netflix['cast'].fillna(value='no cast', inplace=True)
netflix
```

### Lambda Functions
A way of writing a function in one line

Useful for passing a function along many rows

```python
the_adder = lambda x: x + 210101
the_adder(3535)
```

### Map vs Apply
* Map - useful for passing each element of a column to the function
* Apply - useful for passing an entire row to the lambda function

Just a warning, this shit is kinda complicated

```python
# Map
netflix['Genre_count'] = netflix['listed_in'].map(lambda x: len(x.split(', ')))
netflix
```

```python
# Apply
netflix['nonsensical_columns'] = netflix.apply(lambda x: len(x['cast'].split(', ')) + len(x['listed_in'].split(', ')), axis = 1)
netflix
```

### GroupBy, Aggregation, and Sorting Values
Getting the number of media by country

```python
netflix_movies_by_country = netflix.groupby('country')['show_id'].count().reset_index()
netflix_movies_by_country.sort_values(by='show_id',ascending=False)
```

### Unions and Joins
Unions - Vertical merging / Stacking

```python
second_dataset = pd.read_csv(cwd + '/netflix_titles_second.csv')
new_dataset = pd.concat([netflix, second_dataset])
new_dataset
```

Joins - Horizontal merging / The hard one

```python
# Joined_data = pd.merge(left= , right= , how= , left_on=, right_on=) Standard formula for a join
netflix_merged = pd.merge(left=new_dataset, right=netflix_movies_by_country, how='inner', left_on=['country'], right_on=['country'])
netflix_merged
```

## Pivoting Data / Melting data
### Pivoting
Bringing it from "long" to "wide"

```python
# To explain the function below...
# Make a pivot table from the netflix data where the indices are country, make columns from the types with values from the counts of all our titles.
# Anything not found will be zero (Like if theres no Movies from Croatia) rather than NA
pivot_table = netflix.pivot_table(index='country', columns='type', values='title', aggfunc='count', fill_value=0).reset_index().sort_values(by='Movie',ascending=False)
pivot_table
```

### Melting
Bringing it from "wide" to "long"

```python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})
df
```

```python
df.melt(id_vars=['A'], value_vars=['B'],
        var_name='myVarname', value_name='myValname')
```
